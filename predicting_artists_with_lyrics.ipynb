```python
import requests
import re
import time
import numpy as np
from bs4 import BeautifulSoup
```

```python
URL1 = 'https://www.lyrics.com/artist/Michael-Bubl%C3%A9/554516'
URL2= 'https://www.lyrics.com/artist/Ella-Fitzgerald/6503'
```

```python
headers = {'User-agent': 'Mozilla/5.0 (X11; Linux i686; rv:2.0b10) Gecko/20100101 Firefox/4.0b10'}
```

```python
response_buble = requests.get(url = URL1,headers=headers)
response_ella = requests.get(url = URL2,headers=headers)
```

```python

buble_soup = BeautifulSoup(response_buble.text,"html.parser")
ella_soup = BeautifulSoup(response_ella.text,"html.parser")
```

```python
b_links =buble_soup.find_all('strong')
e_links = ella_soup.find_all('strong')
```

```python
buble_links = []
for link in b_links[3:150]:
    f= BeautifulSoup(str(link),'html.parser')
    x = f.a['href']
    buble_links.append(x)



buble_links
```

```python
buble_list = [x for i, x in enumerate(buble_links) if x[-5] in [y[-5] for y in buble_links[i+1:]] or i == len(buble_links) - 1]
buble_list[0]
```

```python
w = open('buble_links.txt','w')
w.close()
for i in buble_list:
    w=open('buble_links.txt','a')
    w.write(f'{i}\n')
    w.close()
```

```python
ella_links = []
for link in e_links[3:150]:
    f= BeautifulSoup(str(link),'html.parser')
    x = f.a['href']
    ella_links.append(x)
```

```python
ella_list = [x for j, x in enumerate(ella_links) if x[-5] in [y[-5] for y in ella_links[j+1:]] or j == len(ella_links) - 1]
len(ella_list)
```

```python
w = open('ella_links.txt','w')
w.close()
for i in ella_list:
    w=open('ella_links.txt','a')
    w.write(f'{i}\n')
    w.close()
```

```python
buble_lyrics=[]
for link in buble_links[:]:
    lyric_url = 'https://www.lyrics.com' + link
    b_songs_url = requests.get(url = lyric_url, headers=headers)
    y = BeautifulSoup(b_songs_url.text,'html.parser')
    buble_lyr = y.find(id='lyric-body-text')
    if buble_lyr:
        x = re.sub(r"\r\n", ' ', buble_lyr.text)
        buble_lyrics.append(x)
buble_lyrics
```

```python
ella_lyrics=[]
for elink in ella_links:
    lyric_urle = 'https://www.lyrics.com/' + elink
    headers = {'User-agent': 'Mozilla/5.0 (X11; Linux i686; rv:2.0b10) Gecko/20100101 Firefox/4.0b10'}
    e_songs_url = requests.get(url = lyric_urle,headers=headers)
    y = BeautifulSoup(e_songs_url.text,'html.parser')
    ella_lyr=y.find(id='lyric-body-text')
    if ella_lyr:
        x = re.sub(r"\r\n", ' ', ella_lyr.text)
        ella_lyrics.append(x)
       
ella_lyrics
    
```

```python
import pandas as pd
df = pd.DataFrame({'Buble':buble_lyrics[:112], 'ella':ella_lyrics[:112]})
df.to_csv('lyrics.csv', index = False)
```
